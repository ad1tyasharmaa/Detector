```python
import cv2
```

```python
import numpy as np
```

```python

face_cap = cv2.CascadeClassifier("C:/Program Files/Python312/Lib/site-packages/cv2/data/haarcascade_frontalface_default.xml")

video_cap = cv2.VideoCapture(0)

lower_skin = np.array([0, 20, 70], dtype=np.uint8)
upper_skin = np.array([20, 255, 255], dtype=np.uint8)


while True:
    ret, video_data = video_cap.read()
    col = cv2.cvtColor(video_data,cv2.COLOR_BGR2GRAY)
    faces = face_cap.detectMultiScale(
        col,
        scaleFactor = 1.1,
        minNeighbors = 5,
        minSize = (30,30),
        flags = cv2.CASCADE_SCALE_IMAGE
    )
    for (x,y,w,h) in faces:
        cv2.rectangle(video_data,(x,y),(x+w,y+h),(0,255,0),2)
    cv2.imshow("video_live", video_data)
    if not ret:
        print("Failed to capture video")
        break

    hsv_frame = cv2.cvtColor(video_data, cv2.COLOR_BGR2HSV)
    skin_mask = cv2.inRange(hsv_frame, lower_skin, upper_skin)
    skin_segment = cv2.bitwise_and(video_data, video_data, mask=skin_mask)

    faces = face_cap.detectMultiScale(video_data, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    for (x, y, w, h) in faces:
        face_roi = skin_segment[y:y + h, x:x + w]
        average_color = cv2.mean(face_roi, mask=skin_mask[y:y + h, x:x + w])[:3]

        if np.mean(average_color) < 60:
            print("Detected Skin Color: Black")
        elif np.mean(average_color) < 150:
            print("Detected Skin Color: Brown")
        else:
            print("Detected Skin Color: White")

    cv2.imshow("video_live", video_data)

    if cv2.waitKey(10) == ord("a"):
        break

video_cap.release()
cv2.destroyAllWindows()
```
